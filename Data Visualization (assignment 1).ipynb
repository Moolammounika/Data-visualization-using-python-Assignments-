### Que 2

### subplots

```python
x=[1,2,3,4,5,6,7] #days
y1=[160,150,140,145,175,165,180] #sales 1
y2=[70,90,160,150,140,145,175]  #sales2
```

```python
plt.plot(x,y1,'red',label='y1')
plt.plot(x,y2,'green',label='y2')
plt.xlabel('days',color='blue')
plt.ylabel('sales',color='blue')
plt.title('Days vs Sales plot',color='orange')
plt.legend()
```

```python
plt.title('subplot')
plt.subplot(2,2,1)
plt.plot(x,y1,'g--')
plt.subplot(2,2,2)
plt.plot(x,y2,'b--')
```
