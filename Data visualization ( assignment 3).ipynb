### Relation plot

```python
sns.relplot(x="timepoint",y="signal",data=data)
```

```python
#using hue
sns.relplot(x="timepoint",y="signal",data=data,hue='region')
```

```python
sns.relplot(x="timepoint",y="signal",data=data,hue='region',style='event',size='subject')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=data,sort=True,hue='region',style='event')
```

### Box plot

```python
sns.boxplot(x="timepoint",y="signal",data=data,hue='region')
```
